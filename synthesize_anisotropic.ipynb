```python
from functions import *
import pathlib as p
```

```python
#LOAD ISOTROPIC GROUND TRUTH IMAGES

DATA_DIR = p.Path.home()/"data"/"bobsrepository" #path to folder with data
isotropic_images = sorted(DATA_DIR.rglob("*T2w.nii.gz")) 
```

```python
#CREATE ANISOTROPIC IMAGES

scale_factor = 2 # slice thickness in mm
start_slice = 0 # which slices to remove, 0 means even slices are removed, 1 means odd slices are removed
direction = 'axial' # direction can be 'axial', 'sagittal' or 'coronal' depending on which plane you want to remove slices from
LR_DIR = DATA_DIR/"LR_data"

#create directory if it doesn't exist
LR_DIR.mkdir(parents=True, exist_ok=True)

for i in range(len(isotropic_images)):
    img = nib.load(isotropic_images[i])
    anisotropic_image = create_LR_img(img.get_fdata(), scale_factor, start_slice, direction)
    nib.save(nib.Nifti1Image(anisotropic_image, img.affine), LR_DIR / isotropic_images[i].name.replace("T2w",f"T2w_LR{scale_factor}_{direction}"))
```
