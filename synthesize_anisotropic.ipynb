```python
from functions import *
import pathlib as p
```

```python
#LOAD ISOTROPIC GROUND TRUTH IMAGES

DATA_DIR = p.Path.home()/"data"/"bobsrepository"
isotropic_images = sorted(DATA_DIR.rglob("*T2w.nii.gz"))
```

```python
#CREATE ANISOTROPIC IMAGES

scale_factor = 2 # slice thickness in mm
start_slice = 0 # which slices to remove, 0 means even slices are removed, 1 means odd slices are removed
direction = 'axial' # direction can be 'axial', 'sagittal' or 'coronal' depending on which plane you want to remove slices from
LR_DIR = DATA_DIR/"LR_images"

#create directory if it doesn't exist
(LR_DIR/f"{direction}").mkdir(parents=True, exist_ok=True)

for i in range(2):
    img = nib.load(isotropic_images[i]).get_fdata()
    lr_images = create_LR_img(img, scale_factor, start_slice, direction)
    nib.save(nib.Nifti1Image(lr_images, affine=nib.load(isotropic_images[i]).affine), LR_DIR/f"{direction}"/f"{isotropic_images[i].name}_LR{scale_factor}.nii.gz")
```
